# Handle missing values

For categorical data, missing values are replaced with mode (most frequent value) because:

* It preserves data distribution without removing rows.
* It logically fills gaps since the most common category is likely correct.
* The missing percentage for Teacher_Quality is (78/6607) ≈ 1.18%, which is low, so mode imputation is appropriate.

```python
# filling missing value ffor Teacher_Quality column
data["Teacher_Quality"].fillna(data["Teacher_Quality"].mode()[0], inplace=True)

```

```python
# filling missing value ffor Parental_Education_Level column
# since this is also categorical and missing values is 1.36 % , mode is the best option


data["Parental_Education_Level"].fillna(data["Parental_Education_Level"].mode()[0], inplace=True)

```

```python
data["Distance_from_Home"]
```

Since Distance_from_Home is categorical, we cannot use mean/median.
The missing percentage is (67/6607) ≈ 1.01%, which is low.
Filling with mode (most frequent value) ensures consistency.

```python
data["Distance_from_Home"].fillna(data["Distance_from_Home"].mode()[0],inplace=True)
```

```python
# checking for missing  values again
data.isnull().sum()
```

```python
# Boxplots (Check Outliers)

for i in ["Hours_Studied", "Sleep_Hours", "Exam_Score",'Attendance']:
    plt.figure(figsize=(5,3))
    sns.boxplot(x=data[i])
    plt.show()
```

FROM THE ABOVE GRAPHS WE UNDESRSTAND THAT THERE ARE OUTLIERS IN EXAM_SCORES , LETS CHECK THEM WITH IQR METHOD

```python
q1 = data["Exam_Score"].quantile(0.25)
q3 = data["Exam_Score"].quantile(0.75)
iqr = q3-q1
exam_score_outliers = data[(data["Exam_Score"] < (q1 - 1.5 * iqr)) | (data["Exam_Score"] > (q3 + 1.5 * iqr))]
```

```python
exam_score_outliers.shape  # total 104 outliers
```

```python
# LETS SEE THE DISTRIBUTION OF THE EXAM MARKS OF OUTLIERS
sns.boxplot(x=exam_score_outliers["Exam_Score"],color="lightblue")
plt.title("Boxplot of Exam Scores of outliers ")
plt.xlabel("Exam Score")
plt.show()
```

# Insights
* FROM THE ABOVE GRAPHS WE UNDESRSTAND THAT THERE ARE OUTLIERS IN EXAM_SCORES.
* there are 104 entries who are considered as outliers
* the scores less than 60 and more than 80
